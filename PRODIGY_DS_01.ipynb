```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('population.csv',  encoding='utf-8')
data.head(265)
     
```

```python
data.describe
```

```python
#List of contries for the graph.
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2019'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2019')
plt.title('Population of selected countries in 2019')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()
```

```python

data_final = data.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
chosen_country = 'United States'
country_data = data_final[data_final['Country Name'] == chosen_country]
years = country_data.columns[7:].astype(int)
population = country_data.iloc[:, 7:].values.flatten()
plt.figure(figsize=(12, 3))
plt.bar(years, population, color='RED')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.show()
```
